Acto 0:Entrelazamiento

"Correlaciones no locales."
En el sandbox, añade un segundo qubit a tu circuito.

    Intenta construir un Estado de Bell (que los dos qubits den siempre el mismo resultado). Para esto, usa una compuerta H en el primer qubit y una compuerta CNOT (el control sobre el primer qubit y el target sobre el segundo).

    Mira las probabilidades resultantes. ¿Por qué el resultado de medir el Qubit 2 depende instantáneamente del resultado del Qubit 1? Explica este fenómeno mediante la covarianza.

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

def calcular_covarianza(counts, shots):
    """
    Calcula la covarianza entre dos qubits mapeando 0->-1 y 1->+1
    """
    suma_xy = 0
    suma_x = 0
    suma_y = 0

    for outcome, count in counts.items():
        # Los qubits vienen como '01' o '10' (el de la izquierda es qubit 0)
        bit_a = int(outcome[1]) # Qubit 0
        bit_b = int(outcome[0]) # Qubit 1

        # Mapeo: 0 -> -1, 1 -> +1
        val_a = 2 * bit_a - 1
        val_b = 2 * bit_b - 1

        suma_xy += (val_a * val_b) * count
        suma_x += val_a * count
        suma_y += val_b * count

    exy = suma_xy / shots
    ex = suma_x / shots
    ey = suma_y / shots

    return exy - (ex * ey)

# 1. Creamos el circuito: Estado de Bell |Phi+>
qc = QuantumCircuit(2)
qc.h(0)        # Superposición
qc.cx(0, 1)    # Entrelazamiento
qc.measure_all()

# 2. Simulamos
simulator = AerSimulator()
shots = 2000
job = simulator.run(transpile(qc, simulator), shots=shots)
counts = job.result().get_counts()

# 3. Calculamos
cov = calcular_covarianza(counts, shots)

print(f"Resultados de medición: {counts}")
print(f"Covarianza calculada: {cov:.4f}")
```
